```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_excel("data/INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls")
```

### Plotting department wise performances

```python
plt.figure(figsize=(14,7))
sns.countplot(x="EmpDepartment",hue="PerformanceRating",palette="pastel",data=data)
plt.show()
```

```python
data_new = data.groupby(["EmpDepartment","PerformanceRating"]).size().reset_index(name="counts")
```

```python
print(data_new)
```

```python

data.groupby(["EmpDepartment"]).agg(['mean','count'])['PerformanceRating'].sort_values('mean', ascending=False)
```
